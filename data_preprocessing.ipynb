```python
from skimage import io
import os
image_dir='datasets/jinnan/normal/'
```

```python
images_path=os.listdir(image_dir)
```

```python
images_path
```

```python
for x in range(len(images_path)):
    image={'coco_url': '',
            'data_captured': '',
             'file_name':'' ,
             'flickr_url': '',
             'id': 980,
             'height': 306,
             'width': 387,
             'license': 1}
    img=io.imread(os.path.join(image_dir,images_path[x]))
    image['file_name']=images_path[x]
    image['id']=981+x
    image['height']=img.shape[0]
    image['width']=img.shape[1]
    data['images'].append(image)
    
```

```python
len(data['images'])
```

划分验证集与训练集，验证集设为100张

```python
np.random.seed(32)
index = np.random.permutation(len(data['images']))
```

```python
permuted_images=[data['images'][x] for x in index]
```

```python
training_images=permuted_images[500:]
val_images=permuted_images[:500]
```

```python
print(len(training_images))
print(len(val_images))
```

```python
training_images
```

```python
trining_id=[]
for image in training_images:
    trining_id.append(image['id'])
    
```

```python
trining_id
```

```python
annotations=data['annotations']
len(annotations)
```

```python
annotations
```

```python
for an in annotations:
    an['area']=an['bbox'][2]*an['bbox'][3]
```

```python
annotations
```

```python
train_annotation=[]
val_annotation=[]

for annotation in annotations:
    image_id=annotation['image_id']
    if image_id in trining_id:
        train_annotation.append(annotation)
    else:
        val_annotation.append(annotation)
    
    
```

统计验证集与训练集的标注框个数

```python
len(train_annotation)
```

```python
len(val_annotation)
```

构建验证集和训练集数据结构

```python
train_data={}
train_data['info']=data['info']
train_data['licenses']=data['licenses']
train_data['categories']=data['categories']
train_data['images']=training_images
train_data['annotations']=train_annotation
```

```python
val_data={}
val_data['info']=data['info']
val_data['licenses']=data['licenses']
val_data['categories']=data['categories']
val_data['images']=val_images
val_data['annotations']=val_annotation
```

写入json文件

```python
with open("./train.json",'w',encoding='utf-8') as json_file:
    json.dump(train_data,json_file,ensure_ascii=False)
```

```python
with open("./val.json",'w',encoding='utf-8') as json_file:
    json.dump(val_data,json_file,ensure_ascii=False)
```
