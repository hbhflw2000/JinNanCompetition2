```python
import pandas as pd
import os
import json
import numpy as np
from skimage import io
```

```python
images='datasets/jinnan/images/'
ann='datasets/jinnan/annotations/train_no_poly.json'
```

```python
label_path=ann
```

```python
print(label_path)
```

```python
f = open(label_path,encoding='utf-8')
data = json.load(f)
```

```python
data
```

```python
data.keys()
```

```python
test_data={}
test_data['info']=data['info']
test_data['licenses']=data['licenses']
test_data['categories']=data['categories']
test_data['images']=[]
```

```python
images_dir='datasets/jinnan/test/'
```

```python
images_path=os.listdir(images_dir)
```

```python
images_path.sort(key = lambda x: int(x[:-4]))
images_path
```

```python
images=[]
for x in range(len(images_path)):
    image={}
    img=io.imread(os.path.join(images_dir,images_path[x]))
    image['file_name']=images_path[x]
    image['id']=x
    image['height']=img.shape[0]
    image['width']=img.shape[1]
    images.append(image)
images
```

```python
test_data['images']=images
```

```python
with open("./test.json",'w',encoding='utf-8') as json_file:
    json.dump(test_data,json_file,ensure_ascii=False)
```
